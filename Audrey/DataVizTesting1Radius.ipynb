```python
import numpy as np
import matplotlib as plt
from pylab import *
%matplotlib inline
```

```python
## Reads the data file but only from the columns in @usecols
data = loadtxt('output.dat', usecols = (1, 2, 3, 10, 11, 13, 14))
#print data

```

```python
## assigns each column in data to the specific value the data represents
bodNum = data[:,0][::101]  # colon equals everything
mass = data[:,1]
time = data[:,2]
xpos = data[:,3]
ypos = data[:,4]
velx = data[:,5]
vely = data[:,6]
r = np.sqrt(xpos**2 + ypos**2)
```

```python
## Need to work on using for loop to generate bodNum
fig = plt.figure()
bodNum = 'bodNum'
for i in range(24):
    obj = bodNum = i
    plt.plot(time[obj],r[obj],label="%d"%i)

#all of the below have to be outside of the for loop so they only happen once
plt.legend(fontsize=15)
plt.xlabel('Time (Days)',fontsize=20)
plt.ylabel('Distance from SMBH',fontsize=20)
plt.gca().tick_params(labelsize=20)
#plt.savefig('25body_pathA.png') #needs to be outside the for loop
```
