```python
import numpy as np
import matplotlib as plt
from pylab import *
%matplotlib inline
```

```python
## reads the data for the specified columns 
data = loadtxt('output.dat', usecols = (1,2,3,10,11))
```

```python
bod_num = data[:,0] [::101]
mass = data[:,1]
time =  data[:,2]
x =  data[:,3]
y =  data[:,4] 
```

```python
#print (bod_num)
print (mass[0])
print (mass[1])
```

```python
index = 0
fig = plt.figure()
for i in bod_num:
    if i == 0:
        colorDot = 'red'
    if i == 1:
        colorDot = 'blue'
    #if i==0 & ((i-1) == 0):
        #colorDot = 'purple'
    plt.plot(x[index], y[index] ,'.', c = colorDot)
    index += 1
    plt.title("Position of black holes over time")
    plt.xlabel("x position")
    plt.ylabel("y position")
    plt.legend()
```
