```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
%matplotlib inline
```

```python
data=loadtxt('output.dat', usecols=(1,2,3,10,11))
```

```python
bod_num=data[:,0]
#for every row in the 0 column is bod_num
#: means every row, column 0
mass=data[:,1]
#for every row in the 1 column is mass
time=data[:,2]
x=data[:,3]
y=data[:,4]
r=np.sqrt(x**2+y**2)
```

```python
fig=plt.figure(figsize = (11,8))#initiates a figure
for i in range(25): #range is a list of numbers from 0 to 24 (so 25 numbers total)
    obj=bod_num==i #boolean indexing
    plt.plot(time[obj],r[obj])
    
#all of the below have to be outside of the for loop so they only happen once
plt.xlabel('Time (Days)',fontsize=20)
plt.ylabel('Distance from SMBH',fontsize=20)
#plt.legend(fontsize=15)
#plt.xlim(250000,500000)
#plt.ylim(665.25,666.50)
plt.gca().tick_params(labelsize=20)
plt.savefig('25body_pathA.png') #needs to be outside the for loop
```
